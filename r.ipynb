В блоке кода сверху мы испортирует библиотеки numpy, pandas и matplotlib.

````python
terraria = pd.read_csv('terraria.csv')
terraria
````

Читаем наш файл с оружиями из игры Terraria, записываем в переменную и выводим в виде таблицы.

````python
# NAME,CLASS,GAME PROGRESSION,DPS (SINGLE TARGET),DPS (MULTI TARGET),DPS (SINGLE TARGET + PROYECTILE ONLY),DPS (MULTI TARGET + PROYECTILE ONLY),OBSERVATIONS

plt.plot(terraria['NAME'].head(4), terraria['DPS (SINGLE TARGET)'].head(4), marker = 'o', markersize = 7, label = 'Одиночный урон', color = 'blue')
plt.plot(terraria['NAME'].head(4), terraria['DPS (MULTI TARGET)'].head(4), marker = 'x', markersize = 7, label = 'Множественный урон', color = 'purple')
plt.legend()
plt.show()
````

Линейный график соотношения одиночного и множественного урона у 4 самых мощных оружий в террарии, с помощью него мы можем наблюдать что у Последней призмы множественный урон меньше, а одиночный высок, а у меча Зенит всё наоборот.
plt.plot создает график, plt.legend показывает легенду а plt.show выводит нам график.

````python
plt.figure().set_figwidth(20)
plt.bar(terraria['NAME'].head(10), terraria['GAME PROGRESSION'].head(10), label = "Прогресс игры", color = 'pink')
plt.plot(terraria['NAME'].head(10), terraria['CLASS'].head(10), marker = "o", markersize = 5, label = 'Класс', color = 'grey' )
plt.legend()
plt.show()
````

В первой строке мы расширяем наше полотно, линией мы смотрим оружие какого класса перед нами, стобцами какой стадии прохождения оружия и видим что больше всего оружий после Лунного Лорда являются оружиями ближнего боя.

````python
plt.figure().set_figwidth(15)
plt.plot(terraria['NAME'].tail(10),terraria['DPS (SINGLE TARGET)'].tail(10), marker = '*', markersize = 5, label = 'Урон', color = 'red')
plt.legend()
plt.show()
````

Это линейный график соотношения урона, касательно цены, снизу подписаны названия оружий, слева написан урон в единицах, справа сверху легенда.

````python
plt.figure().set_figwidth(15)
plt.bar(terraria['GAME PROGRESSION'].head(20),terraria['DPS (MULTI TARGET)'].head(20), label = 'Урон', color = 'grey')
plt.legend()
plt.show()
````

Здесь продемонстрирована столбчатая диаграмма соотношения урона оружия по множественной цели, касательно прогресса в игре. Снизу прогресс, слева урон, справа сверху легенда.

````python
plt.bar(terraria['NAME'].head(5),terraria['DPS (MULTI TARGET)'].head(5), label = 'Урон', color = 'blue', alpha = 0.3)
plt.legend()
plt.show()
````

Это столбчатая диаграмма соотношения уорна, касательно названия. Данные выбираются по первым пяти названиям в файле, столбики диаграммы означают вес, сами они голубые и, благодаря параметру "alpha" полупрозрачные. Маркеры здесь не нужны, т.к. это диаграмма. 

````python
data = terraria['DPS (SINGLE TARGET)'].value_counts().head(6)
plt.pie(data,labels=data.index,autopct='%1.1f%%', colors = ['pink','yellow'])
plt.legend()
plt.show()
````

Это круговая диаграмма, которая содержит в себе процентное соотношение урона оружий. Сначала читается файл, затем данные выбираются по первым шести названиям в файле, заполняется список информации с помощью value_counts(). Autopct='%1.1f%%' значит, что данные будут представлены в процентах. Названия будут подписаны благодаря коллекции.
